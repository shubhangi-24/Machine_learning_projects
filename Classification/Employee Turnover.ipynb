Train Test Split

```python
x=df.drop(columns=['Attrition (Yes/No)'],axis=1)
y=df['Attrition (Yes/No)']
```

```python
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
```

```python
# Model 1
# Using Logistic Regression

from sklearn.linear_model import LogisticRegression
clf1=LogisticRegression()
clf1.fit(x_train,y_train)
y_test_pred=clf1.predict(x_test)
y_train_pred=clf1.predict(x_train)

from sklearn.metrics import accuracy_score
train_acc=accuracy_score(y_train,y_train_pred)
test_acc=accuracy_score(y_test,y_test_pred)
print('Training Acc=',train_acc*100)
print('Testing Acc=',test_acc*100)
```

```python
from sklearn.metrics import confusion_matrix
cm1=confusion_matrix(y_test,y_test_pred)
print(cm1)

sns.heatmap(cm1,annot=True,cbar=False,fmt='.1f')
```

```python
# Model 2 
# Using Decision Tree

from sklearn.tree import DecisionTreeClassifier
clf2=DecisionTreeClassifier()
clf2.fit(x_train,y_train)
y_test_pred=clf2.predict(x_test)
y_train_pred=clf2.predict(x_train)

from sklearn.metrics import accuracy_score
train_acc=accuracy_score(y_train,y_train_pred)
test_acc=accuracy_score(y_test,y_test_pred)
print('Training Acc=',train_acc*100)
print('Testing Acc=',test_acc*100)
```

```python
from sklearn.metrics import confusion_matrix
cm2=confusion_matrix(y_test,y_test_pred)
print(cm2)

sns.heatmap(cm2,annot=True,cbar=False,fmt='.1f')
```

```python
# Model 3
# Using SVM

from sklearn.svm import SVC
clf3=SVC()
clf3.fit(x_train,y_train)
y_test_pred=clf3.predict(x_test)
y_train_pred=clf3.predict(x_train)

from sklearn.metrics import accuracy_score
train_acc=accuracy_score(y_train,y_train_pred)
test_acc=accuracy_score(y_test,y_test_pred)
print('Training Acc=',train_acc*100)
print('Testing Acc=',test_acc*100)
```

```python
from sklearn.metrics import confusion_matrix
cm3=confusion_matrix(y_test,y_test_pred)
print(cm3)

sns.heatmap(cm3,annot=True,cbar=False,fmt='.1f')
```

```python
# Model 4
# Using KNN

from sklearn.neighbors import KNeighborsClassifier
clf4= KNeighborsClassifier()
clf4.fit(x_train,y_train)
y_test_pred=clf4.predict(x_test)
y_train_pred=clf4.predict(x_train)
from sklearn.metrics import accuracy_score
train_acc=accuracy_score(y_train,y_train_pred)
test_acc=accuracy_score(y_test,y_test_pred)
print('Training Acc=',train_acc*100)
print('Testing Acc=',test_acc*100)
```

```python
from sklearn.metrics import confusion_matrix
cm4=confusion_matrix(y_test,y_test_pred)
print(cm4)

sns.heatmap(cm4,annot=True,cbar=False,fmt='.1f')
```

```python
# Model 5
# Using RandomForest Classifier

from sklearn.ensemble import RandomForestClassifier
clf5=RandomForestClassifier()
clf5.fit(x_train,y_train)
y_test_pred=clf5.predict(x_test)
y_train_pred=clf5.predict(x_train)

from sklearn.metrics import accuracy_score
train_acc=accuracy_score(y_train,y_train_pred)
test_acc=accuracy_score(y_test,y_test_pred)
print('Training Acc=',train_acc*100)
print('Testing Acc=',test_acc*100)
```

```python
from sklearn.metrics import confusion_matrix
cm5=confusion_matrix(y_test,y_test_pred)
print(cm5)

sns.heatmap(cm5,annot=True,cbar=False,fmt='.1f')
```

Comparison of different models

```python
#  Compare the performance of different classifier

clf=[clf1,clf2,clf3,clf4,clf5]
name=['LR','DT','SVM','KNN','RF']
accuracy={}
for model,model_name in zip(clf,name):
    model.fit(x_train,y_train)
    acc=accuracy_score(model.predict(x_test),y_test)
    accuracy[model_name]=acc*100
for i,j in accuracy.items():
    print(i,':',j)
    
classifiers=[]
ac=[]
for i,j in accuracy.items():
    classifiers.append(i)
    ac.append(j)

fig = plt.figure(figsize = (10, 5))
#  Bar plot
plt.bar(classifiers,ac,color="Black")
plt.xlabel("models")
plt.ylabel("accuracy")
plt.title("accuracy of different models")
plt.show()
```
